```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.databricks.com/dataaisummit/speakers?itm_data=hp-featured-speakers-dais23"  # Replace with the URL of the website you want to scrape
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#print(soup.prettify())
```

```python
speaker_names = soup.find_all(class_="css-78idzx")
```

```python
names=[]
```

```python
for element in speaker_names:
    print(element.text)
    names.append(element.text)
```

```python
import pandas as pd

df = pd.DataFrame(names, columns=['Names'])
print(df)
```

```python
df.to_csv('Names_of_the_Speakers.csv', index=False)
```
